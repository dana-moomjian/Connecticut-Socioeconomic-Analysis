# **Data Cleaning and Merging**

**Datasets Printed and Columns Renamed Prior to Merge**

```python
import pandas as pd
children_df = pd.read_csv('/content/Children_by_Family_Type_2164570551476242135.csv')
educational_df = pd.read_csv('/content/Educational_Attainment_8867652791947649948.csv')
employment_df = pd.read_csv('/content/Employment_by_Occupation_7163685701367074130.csv')

children_df = children_df.rename(columns={
    'Family Type': 'Children Family Type',
    'Value': 'Children Value',
    'Margin of Error': 'Children Margin of Error',
    'Measure': 'Children Measure Type',
    'GeoID': 'GeoID',
    'Geography Name': 'Town',
    'ObjectId': 'ObjectId'})

educational_df = educational_df.rename(columns={
    'Sex': 'Education Sex',
    'Education': 'Education Level',
    'Value': 'Education Value',
    'Margin of Error': 'Education Margin of Error',
    'Measure': 'Education Measure Type',
    'GeoID': 'GeoID',
    'Geography Name': 'Town',
    'ObjectId': 'ObjectId'})

employment_df = employment_df.rename(columns={
    'Occupation': 'Employment Occupation',
    'Value': 'Employment Value',
    'Margin of Error': 'Employment Margin of Error',
    'Measure': 'Employment Measure Type',
    'GeoID': 'GeoID',
    'Geography Name': 'Town',
    'ObjectId': 'ObjectId'})


print("Children by Family Type (renamed columns):")
display(children_df.head())

print("\nEducational Attainment (renamed columns):")
display(educational_df.head())

print("\nEmployment by Occupation (renamed columns):")
display(employment_df.head())
```

**Removing Duplicates in the Data**

```python
children_df.drop_duplicates(inplace=True)
educational_df.drop_duplicates(inplace=True)
employment_df.drop_duplicates(inplace=True)

print("Shape of children_df after removing duplicates:", children_df.shape)
print("Shape of educational_df after removing duplicates:", educational_df.shape)
print("Shape of employment_df after removing duplicates:", employment_df.shape)
```

**Merged Data After Cleaning**

```python

children_town_df = children_df[children_df['Geography Type'] == 'Town'].copy()
educational_town_df = educational_df[educational_df['Geography Type'] == 'Town'].copy()
employment_town_df = employment_df[employment_df['Geography Type'] == 'Town'].copy()

merged_df = pd.merge(children_town_df, educational_town_df, on=['Year', 'GeoID', 'Geography Type', 'Town'], how='inner')
merged_df = pd.merge(merged_df, employment_town_df, on=['Year', 'GeoID', 'Geography Type', 'Town'], how='inner')

display(merged_df.sample(5))
```

# **Summary Tables Highlighting Socioeconomic Trends Across Connecticut Towns**

**Table 1: Top 10 Towns by Total Children**


```python
latest_year = merged_df['Year'].max()
children_total_latest_year = merged_df[(merged_df['Year'] == latest_year) & (merged_df['Children Family Type'] == 'Total')]

top_10_children_towns = children_total_latest_year.groupby('Town')['Children Value'].sum().reset_index()

top_10_children_towns = top_10_children_towns.sort_values(by='Children Value', ascending=False).head(10)
display(top_10_children_towns[['Town', 'Children Value']])
```

Based on the data, the ten Connecticut towns with the highest number of children are Bridgeport, New Haven, Hartford, Waterbury, Stamford, Greenwich, Norwalk, Danbury, Fairfield, and New Britain. These towns represent a mix of Connecticut’s largest urban centers and some affluent suburban areas. Bridgeport leads by a significant margin, followed by New Haven and Hartford, which aligns with their status as major population hubs with large family communities. Interestingly, towns like Greenwich and Fairfield who are known for their higher median incomes also appear on the list, suggesting that both urban and suburban areas contribute heavily to the state’s child population. This distribution could imply varying needs in education, childcare, and community resources across socioeconomic lines.

**Table 2: Distribution of Educational Attainment Levels in a Sample Town**



```python

latest_year = merged_df['Year'].max()
latest_year_children = merged_df[(merged_df['Year'] == latest_year) & (merged_df['Children Family Type'] != 'Total')].copy()

average_children_by_family_type = latest_year_children.groupby('Children Family Type')['Children Value'].agg(['mean', 'std']).reset_index()

average_children_by_family_type = average_children_by_family_type.rename(columns={'mean': 'Average Children Value', 'std': 'Children Value Standard Deviation'})

display(average_children_by_family_type)
```

Looking at the data, clear trends emerge in family size by household type. Married couples have the highest average number of children, which may reflect traditional family structures where two adults share resources, time, and responsibilities, making it easier to raise more children. Female householders without a spouse have a moderate average number of children, but the very high standard deviation suggests that while some women in this group manage larger families, many have only one or no children possibly due to economic constraints, work demands, or limited support systems. Male householders without a spouse have the lowest average number of children and a relatively low variation, which could indicate that single fathers are less likely to have large families.

**Table 3: Employment by Occupation in a Sample Town**

```python
sample_towns = ['Greenwich', 'Waterbury']
latest_year = merged_df['Year'].max()
latest_year_data = merged_df[(merged_df['Town'].isin(sample_towns)) &
                             (merged_df['Year'] == latest_year)].copy()

children_total_latest_year_sample = latest_year_data[latest_year_data['Children Family Type'] == 'Total'].groupby('Town')['Children Value'].sum().reset_index()
children_total_latest_year_sample = children_total_latest_year_sample.rename(columns={'Children Value': 'Total Children'})

employment_sample_towns = latest_year_data[latest_year_data['Employment Occupation'] != 'Total']
employment_comparison_grouped = employment_sample_towns.groupby(['Town', 'Employment Occupation'])['Employment Value'].sum().reset_index()

employment_comparison_pivot = employment_comparison_grouped.pivot(index='Employment Occupation', columns='Town', values='Employment Value').reset_index()

employment_reshaped = employment_comparison_grouped.pivot(index='Town', columns='Employment Occupation', values='Employment Value').reset_index()

comparison_table = pd.merge(employment_reshaped, children_total_latest_year_sample, on='Town', how='left')

comparison_table['Year'] = latest_year
display(comparison_table)
```

This data compares Greenwich, one of Connecticut’s wealthiest towns, with Waterbury, a more working class community. The contrast highlights clear economic and occupational differences. Greenwich has the majority of its workforce in management, business, science, and arts (high paying, white collar sectors) while Waterbury has more workers in service, production, and construction roles, which are typically lower income fields. Despite Greenwich’s wealth, it has fewer total children than Waterbury, suggesting that affluent areas may have smaller family sizes and fewer young residents. Waterbury’s higher number of children reflects a larger working class population with more families and dependents. This comparison emphasizes how income and job type influence family structure and population patterns across Connecticut towns.

# **Comparative Plots Highlighting Socioeconomic Trends Across Connecticut Towns**

**Plot 1: Distribution of Educational Attainment Levels in Connecticut Towns**

```python
import matplotlib.pyplot as plt
import seaborn as sns
latest_year = merged_df['Year'].max()
latest_year_education = merged_df[(merged_df['Year'] == latest_year) & (merged_df['Education Level'] != 'Total')].copy()
education_distribution = latest_year_education.groupby('Education Level')['Education Value'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='Education Level', y='Education Value', data=education_distribution)
plt.title('Distribution of Educational Attainment Levels in Connecticut Towns (Latest Year)')
plt.xlabel('Educational Attainment Level')
plt.ylabel('Total Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

This bar chart shows the distribution of educational attainment levels across Connecticut towns. The data reveals that the largest group of residents hold a bachelor’s degree or higher, indicating a generally well-educated population statewide. The next largest group consists of high school graduates or equivalents, followed by those with some college experience. Smaller portions of the population have associate’s degrees or less than a high school diploma. This pattern suggests that Connecticut has a strong base of higher education, which likely contributes to its strong employment rates in professional and technical occupations. It also highlights how education may play a key role in shaping economic opportunities and quality of life across the state.

 **Plot 2: Total Children vs. Total Employment in Connecticut Towns**

```python
latest_year = merged_df['Year'].max()
latest_year_totals = merged_df[(merged_df['Year'] == latest_year) &
                               (merged_df['Children Family Type'] == 'Total') &
                               (merged_df['Employment Occupation'] == 'Total')].copy()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Children Value', y='Employment Value', data=latest_year_totals)
plt.title('Total Children vs. Total Employment in Connecticut Towns (Latest Year)')
plt.xlabel('Total Children')
plt.ylabel('Total Employment')
plt.grid(True)
plt.show()
```

This scatter plot shows the relationship between total employment and total children across towns in Connecticut. The data reveals a positive correlation where as total employment increases, the number of children in a town also tends to rise. This suggests that towns with stronger job markets and more employment opportunities are also likely to support larger family populations. However, there is noticeable variation among towns with lower employment and child counts, likely reflecting smaller or less economically active communities. The upward trend highlights how economic stability and family size may be interconnected, with towns offering more jobs often being more attractive for families with children.

 **Plot 3: Employment Based on Occupation in Selected Connecticut Towns**

```python
latest_year = merged_df['Year'].max()
selected_towns = ['Hartford', 'Stamford', 'Bridgeport']
latest_year_employment_selected_towns = merged_df[(merged_df['Town'].isin(selected_towns)) &
                                                  (merged_df['Year'] == latest_year) &
                                                  (merged_df['Employment Occupation'] != 'Total')].copy()

plt.figure(figsize=(12, 7))
sns.barplot(x='Employment Occupation', y='Employment Value', hue='Town', data=latest_year_employment_selected_towns)
plt.title('Employment by Occupation in Selected Connecticut Towns (Latest Year)')
plt.xlabel('Occupation Category')
plt.ylabel('Total Value')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Town')
plt.tight_layout()
plt.show()
```

This bar chart compares employment by occupation in the Connecticut towns of Hartford, Bridgeport, and Stamford. The data shows that Stamford has a much higher concentration of workers in management, business, science, and arts occupations, reflecting its status as a professional and corporate hub. Bridgeport, on the other hand, has more workers in service and production-related fields, aligning with its more industrial and working-class economy. Hartford shows a more balanced distribution across occupational categories but generally lower employment totals than Stamford or Bridgeport. Overall, the chart highlights how economic diversity and occupational structure vary significantly among Connecticut towns, which can impact income levels, family demographics, and access to resources.
